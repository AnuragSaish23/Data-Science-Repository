# Pandas - Series and DataFrames
- It is a powerful Python Library mainly used for data analysis and data cleaning. It provides 2 Primary data structures : **Series & DataFrames**.
- Series : One - Dimensional array like structure
- DataFrames : 2-D, Size mutable, and potentially a heterogenous tabular Data Structure with labelled axes (Rows, Cols)

```python
import pandas as pd
```

## Series:

```python
# Series from an array 
# This prints the default index of the series which is 0,1,2,3 and the values of the series which are 1,2,3,4

data = [1,2,3,4]

series = pd.Series(data)
print('Series: \n', series)
```

```python
# Series from a Dictionary
# This prints keys as the index of the series which is a,b,c,d and the values of the series which are 1,2,3,4

data = {'a':1, 'b':2, 'c':3, 'd':4}
series_dict = pd.Series(data)

print('Series from a Dictionary: \n', series_dict)
```

```python
# Creating Custom Indexes

data = [1,2,3,4]
index = ['a', 'b', 'c', 'd']

pd.Series(data, index = index)
```

## DataFrames:

```python
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David'],
    'Age' : [25, 30, 35, 40],
    'City' : ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

## Removing a Column from the DataFrame
- the code below is throwing an error because the drop() method does not modify the original dataframe by default. We have to set axis = 1 for it to check the col indexes otherwise by default i.e axis = 0 it check the row indexes and it will not have the col salary hence it is throwing the error.

- The code below will drop the Salary column from the dataframe but it will not change the original dataframe. If you want to change the original dataframe, you can use the inplace parameter and set it to True. For example, df.drop('Salary', inplace=True) will drop the Salary column from the dataframe and change the original dataframe.


```python
df.drop('Salary', axis=1)
```

```python
# The fix
print(df)
df.drop('Salary', axis =1, inplace = True)
df
```
