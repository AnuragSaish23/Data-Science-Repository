# Data Visualization with Seaborn:
- Seaborn is a visualization library based on Matplotlib that provides high level interface for drawing attractive and informative statistical graphics. 
- Seaborn helps in creating complex visualizations with few lines of code.

```python
import seaborn as sns
```

## Basic Plotting with Seaborn

```python
tips = sns.load_dataset('tips')
tips
```

## Creating a scatter plot:

```python
import matplotlib.pyplot as plt

sns.scatterplot(data=tips, x = 'total_bill', y = 'tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

## Creating Line Plot:

```python
sns.lineplot(data = tips, x = 'size', y = 'total_bill')
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

## Creating Plots for Categorical variables:
- Bar Plot

```python
sns.barplot(data = tips, x = 'day', y = 'total_bill')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

## Box Plots: 

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.show()  
```

## Histogram Plot:
- **kde** = kernel density estimation, which adds a smooth curve to the histogram to show the distribution of the data.
- **kernel density estimation (KDE)** is a non-parametric way to estimate the probability density function of a random variable. It provides a smooth curve that represents the distribution of the data, allowing us to visualize the underlying pattern and identify any peaks or modes in the data.

```python
sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

## KDE Plot: 

```python
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
sns.pairplot(tips)
```

## Heatmap:

```python
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```
