```python
### Read a Whole File

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
### read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
### Writing a file(Overwriting) :

with open('example.txt', 'w') as file:
    file.write("Hello World \n")
    file.write('this is a new line')
```

```python
### Writing a file (Without Ovewriting)
with open('example.txt','a') as file:
    file.write('\nThe append fn is being used\n')
    file.write('Give Placement')
```

```python
### Writing a list of lines in a file 
lines = ['First Line \n','Second Line \n','Third Line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
### Binary Files
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
### Reading Binary Files

with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from a Source txt file and write it to the destination text file

# Reading from the Source Text File
with open('example.txt','r') as file:
    content = file.read()

# Writing in the Destination Text File 
with open('destination.txt','w') as file:
    file.write(content)
```

```python
### Read a text file and count the number of lines , words and chars 
def count_text(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, chars = count_text(file_path)
print(f'Lines : {lines}, Words : {words}, Characters : {chars}')
```

```python
### Writing and then Reading a File 
with open('example.txt','w+') as file:
    file.write('My name is Anurag Saish\n')
    file.write('Hello World \n')

    file.seek(0)

    content = file.read()
    print(content)
```

```python
### Reading A File 
def read_file(file_path1):
    with open(file_path1,'r') as file:
        for line in file:
            print(line.strip())
    
read_file('example.txt')
```

```python
def write_file(lines,filename):
    with open(filename,'w') as file:
        for line in lines:
            file.write(line + '\n')
        
write_file(['hello','my name is anurag'],'output.txt')
```

```python
def read_copy(sfilename,dfilename):
    with open(sfilename,'r') as file:
        items = file.read()

    with open(dfilename, 'w') as file:
        file.write(items)

read_copy('output.txt','des.txt')
```

```python
def append_file(lines, filename):
    with open(filename,'a') as file:
        for line in lines:
            file.write(line + '\n')

append_file(['this is prac','so many questions'],'log.txt')
```

```python
def count_words(filename):
    with open(filename,'r') as file:
        lines = file.readlines()
        word_count = sum(len(line.split()) for line in lines)
    return word_count

filename = 'log.txt'
word_count = count_words(filename)
print(f'Words : {word_count}')
```

```python
def findreplace(file_path,old_word,new_word):
    with open(file_path,'r') as file:
        text = file.read()
    new_word = text.replace(old_word,new_word)
    with open(file_path,'w') as file:
        file.write(new_word)

findreplace('log.txt','questions','anurag')
```

```python
def revread(filename):
    with open(filename , 'r') as file:
        lines = file.readlines()
        for line in reversed(lines):
            print(line.strip())

revread('log.txt')
```
