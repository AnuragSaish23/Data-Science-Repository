```python
## Modelling a Bank Account

class BankAccount:
    def __init__(self, ownerinfo, accountNo, balance = 0):
        self.ownerinfo = ownerinfo
        self.__balance = balance
        self.__accountNo = accountNo

    def deposit(self,amount):
        self.__balance += amount
        return f'Account Balance after Deposit : {self.__balance}'
    
    def withdraw(self,amount):
        if amount > self.__balance:
            return 'Insuffucient Funds'
        else:
            self.__balance -= amount
            return f'Account Balance after withdrawal : {self.__balance}'
        
    def get_balance(self):
        return f'The Balance in your account is : {self.__balance}'
    def account_number(self):
        return f'Account Number : {self.__accountNo}'
    
#Creating an Account

acc = BankAccount('Anurag',10000)
print(acc.balance)
acc.deposit(5000)
print(acc.balance)
acc.withdraw(2000)
print(acc.balance)
        
```

```python
# Creating a Car Class

class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        return f'{self.make} engine has started.'

audi = Car('Audi','A8','2020')
print(f'Make : {audi.make}, Model : {audi.model}, Year : {audi.year}')
print(audi.start_engine())
```

```python
# Class Inheritance:
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
class Employee(Person):
    def __init__(self,name,age,employee_id):
        super().__init__(name,age)
        self.employee_id = employee_id
    def __str__(self):
        return f'Employee : [Name : {self.name}, Age : {self.age}, Employee ID : {self.employee_id}]'
emp1 = Employee('Anurag',21,'EA21')
print(emp1)
#print(emp1.name, emp1.age, emp1.employee_id)
```

```python
class Address:
    def __init__(self,street,city,zipcode):
        self.street = street
        self.city = city 
        self.zipcode = zipcode
class Person(Address):
    def __init__(self,name,age, street, city, zipcode):
        super().__init__(street, city, zipcode)
        self.name = name
        self.age = age

person1 = Person('Anurag',21,'Street 10','Dwarka',110078)
print(f'Address : [Name : {person1.name}, Age : {person1.age}, Street : {person1.street}, City : {person1.city}, Zipcode : {person1.zipcode}]')
```

```python
class Address:
    def __init__(self,street,city,zipcode):
        self.street = street 
        self.city = city 
        self.zipcode = zipcode
class Person:
    def __init__(self,name,age,address):
        self.name = name 
        self.age = age
        self.address = address

address = Address('123 Street','Dwarka',110078)
pers1 = Person('Anurag',21,address)

print(pers1.address.street,pers1.address.city,pers1.address.zipcode)
```

```python
class Count:
    count = 0
    def __init__(self):
        Count.count += 1
    @classmethod
    def get_count(cls):
        return cls.count
    
cl1 = Count()
cl2 = Count()
cl3 = Count()
print(Count.get_count())
```

```python
class MathOperations:
    
    @staticmethod
    def sqroot(num):
        return num**0.5
    
print(MathOperations.sqroot(4))
```

```python
class Rectangle:
    def __init__(self,length, width):
        self.__length = length
        self.__width = width

    @property
    def length(self):
        return self.__length
    
    @length.setter
    def length(self, length):
        self.__length = length

    @property
    def width(self):
        return self.__width
    
    @width.setter
    def width(self,width):
        self.__width = width

rect = Rectangle(10,5)
print(rect.length, rect.width)
rect.length = 15
print(rect.length)
```
