```python
def handleZero(a,b):
    try:
        res = a/b
        print(f'The res of the division is : {res}')
    except Exception as ex:
        print(ex)
    finally:
        print('Execution Complete')

handleZero(4,0)
```

```python
def fileOpen(file):
    try:
        file = open('data.txt','r')
    except FileNotFoundError as ex:
        print(ex)
    finally:
        print('File Reading Complete.')
fileOpen('data.txt')
```

```python
def retsum(lis):
    try:
        sumnum = sum(lis)
        print(f'The sum of the elements of the list is : {sumnum}')
    except TypeError as ex:
        print('Please input the correct data type (int/float)')
    except Exception as ex1:
        print(ex1)
    finally:
        print("Calc of sum of the list elements complete.")
retsum([1,2,3,4,'a'])
```

```python
def get_dict_value(d,key):
    try:
        value = d[key]
    except ValueError as ex:
        print(f'Error : {ex}')
    finally:
        print('Done')
    return value
d = {x:x**2 for x in range(1,11)}
print(get_dict_value(d,10)) 
```

```python
def nested(a,b):
    try:
        if type(a) == str:
            a = int(a)
        elif type(b) == str:
            b = int(b)
        try:
            res = a/b
            print(res)
        except ZeroDivisionError as ex:
            print(f'Error : {ex}')
    except ValueError as ex2:
        print(f'Error : {ex2}')
    finally:
        print("Code Execution Complete.")
nested('2',4)
```

```python
def listexception(l,i):
    try:
        value = l[i]
        return value
    except IndexError as e:
        print(f"Error : {e}")
    finally:
        print("Code Execution Complete.")
listexception([x for x in range(1,11)],11)
```

```python
import requests
def urlcheck(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        print(f'Network Error : {e}')
        return None
    finally:
        print("Execution Complete.")

urlcheck('https://nonexistent.url')
```

```python
import json
def jsonhandling(json_string):
    try:
        data = json.loads(json_string)
        return data
    except json.JSONDecodeError as e:
        print(f"JSON ERROR : {e}")
        return None
    finally:
        print("Execution Complete.")
jsonhandling('{"name" : "Anurag" , "Age" : 21}')
```

```python
class NegativeNumberException(Exception):
    pass
def check_for_neg(lst):
    try:
        for num in lst:
            if num < 0:
                raise NegativeNumberException(f'Negative Number Encountered : {num}')
    except NegativeNumberException as e:
        print(f'Error : {e}')
    finally:
        print('Execution Complete')

check_for_neg([1,2,3,-5,6,7])
    
```

```python
class big:
    def method(self,a,b):
        try:
            res = a/b
        except ZeroDivisionError as e:
            print(f'Error : {e}')
            res = None
        finally:
            print('Code Execution Complete')
        return res

calc = big()
print(calc.method(8,2))
```

```python
def convert_integers(lst):
    integers = []
    try:
        for num in lst:
            integers.append(int(num))
    except ValueError as e:
        print(f'Error : {e}')
        integers = None 
    finally:
        print('Code Execution Complete')
    return integers

print(convert_integers(['1','2','3']))
print(convert_integers(['1','Two','3']))
```

```python
def listcompre(lst):
    try:
        integers = [int(x) for x in lst]
    except ValueError as e:
        print(f'Error : {e}')
        integers = None
    finally:
        print('Code Execution Complete')
    return integers
print(listcompre(['1','2','3']))
print(listcompre(['1','2','Three']))
```

```python
def write(strings,filename):
    try:
        file = open(filename,'w')
        for string in strings:
            file.write(string + '\n')
    except IOError as e:
        print(f'Error : {e}')
    finally:
        try:
            file.close()
        except NameError:
            pass

write(['Hello',"World"],'output.txt')
```
